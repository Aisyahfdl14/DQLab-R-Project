# Memanggil Library

```r
library(dplyr)
library(scales)
library(ggplot2)
```

# Import Data

```r
#Data yang digunakan : 'https://storage.googleapis.com/dqlab-dataset/loan_disbursement.csv'

df_loan <- read.csv('https://storage.googleapis.com/dqlab-dataset/loan_disbursement.csv', stringsAsFactors = F)

glimpse(df_loan)
```

```r
#Menampilkan data

df_loan
```

# Memfilter data bulan Mei 2020, dan jumlahkan data per cabang

```r
df_loan_mei <- df_loan %>% 
  filter(tanggal_cair >= '2020-05-01', tanggal_cair <= '2020-05-31') %>% 
  group_by(cabang) %>% 
  summarise(total_amount = sum(amount))

df_loan_mei
```

# Menampilkan data 5 cabang dengan total amount paling besar

```r
df_loan_mei %>% 
  arrange(desc(total_amount)) %>% 
  mutate(total_amount = comma(total_amount)) %>% 
  head(5)
```

# Menampilkan data 5 cabang dengan total amount paling kecil

```r
df_loan_mei %>% 
  arrange(total_amount) %>% 
  mutate(total_amount = comma(total_amount)) %>% 
  head(5)
```

# Menghitung umur cabang (dalam bulan)

```r
df_umur_cabang<- df_loan %>%
  group_by(cabang) %>% 
  summarise(pertama_cair = min(tanggal_cair)) %>% 
  mutate(umur = as.numeric(as.Date('2020-05-15') - as.Date(pertama_cair)) %/% 30) 
df_umur_cabang
```

# Menghitung Umur Cabang (dalam hari)

```r
df_umur_cabang_hari<- df_loan %>%
  group_by(cabang) %>% 
  summarise(pertama_cair = min(tanggal_cair)) %>% 
  mutate(umur = as.numeric(as.Date('2020-05-15') - as.Date(pertama_cair))) 
df_umur_cabang_hari
```

# Menggabungkan data umur dan performa mei

```r
df_loan_mei_umur <- df_umur_cabang %>%
  inner_join(df_loan_mei, by = 'cabang')
df_loan_mei_umur
```

# Plot relasi umur dan performa mei

```r
ggplot(df_loan_mei_umur, aes(x = umur, y = total_amount)) +
  geom_point() +
  scale_y_continuous(labels = scales::comma) +
  labs(title = "Performa Cabang Berdasarkan Umur",
       x = "Umur (bulan)",
       y = "Total Amount", 
      subtitle = "Semakin berumur performa cabang semakin baik") +
  theme(plot.title = element_text(hjust = 0.5, size = 14, face="bold"),
       plot.subtitle = element_text(hjust = 0.5, size = 10))
```

# Mencari cabang yang performa rendah untuk setiap umur

```r
df_loan_mei_flag <- df_loan_mei_umur %>% 
  group_by(umur) %>% 
  mutate(Q1 = quantile(total_amount, 0.25),
	   Q3 = quantile(total_amount, 0.75),
	  IQR = (Q3-Q1)) %>%
  mutate(flag = ifelse(total_amount < (Q1 - IQR), 'rendah','baik'))

df_loan_mei_flag
```

```r
df_loan_mei_flag %>% 
  filter(flag == 'rendah') %>% 
  mutate_if(is.numeric, funs(comma))
```

# Scatterplot dan beri warna merah pada cabang dengan performa rendah

```r
ggplot(df_loan_mei_flag, aes(x = umur, y = total_amount)) +
  geom_point(aes(color = flag)) +
  scale_color_manual(breaks = c("baik", "rendah"),
                        values=c("blue", "red")) +
  scale_y_continuous(labels = scales::comma) +
  labs(title = "Plot Performa Cabang",
       color = "",
       x = "Umur (bulan)",
       y = "Total Amount") +
  theme(plot.title = element_text(hjust = 0.5))
```

# Menggabungkan umur, performa, dan flag

```r
df_all_mei_flag <- df_loan_mei_flag %>%
    inner_join(df_loan, by = 'cabang') %>%
    filter(tanggal_cair >= '2020-05-01', tanggal_cair <= '2020-05-31') %>% 
    group_by(cabang, umur, flag) %>% 
    summarise(jumlah_hari = n_distinct(tanggal_cair),
            agen_aktif = n_distinct(agen),
            total_loan_cair = n_distinct(loan_id),
            avg_amount = mean(amount), 
            total_amount = sum(amount)) %>% 
    arrange(desc(total_amount)) %>% 
    mutate_if(is.numeric, funs(comma))

df_all_mei_flag
```

# Melihat perbadingan performa cabang dengan umur yang sama

```r
df_loan_mei_flag %>% 
  filter(umur == 3) %>% 
  inner_join(df_loan, by = 'cabang') %>% 
  filter(tanggal_cair >= '2020-05-01', tanggal_cair <= '2020-05-31') %>% 
  group_by(cabang, flag) %>% 
  summarise(jumlah_hari = n_distinct(tanggal_cair),
            agen_aktif = n_distinct(agen),
            total_loan_cair = n_distinct(loan_id),
            avg_amount = mean(amount), 
            total_amount = sum(amount)) %>% 
  arrange(total_amount) %>% 
  mutate_if(is.numeric, funs(comma))
```

# Lihat perbadingan performa agen pada cabang yang rendah

```r
df_loan_mei_flag %>% 
  filter(umur == 3, flag == 'rendah') %>% 
  inner_join(df_loan, by = 'cabang') %>% 
  filter(tanggal_cair >= '2020-05-01', tanggal_cair <= '2020-05-31') %>% 
  group_by(cabang, agen) %>% 
  summarise(jumlah_hari = n_distinct(tanggal_cair),
            total_loan_cair = n_distinct(loan_id),
            avg_amount = mean(amount), 
            total_amount = sum(amount)) %>% 
  arrange(total_amount) %>% 
  mutate_if(is.numeric, funs(comma))
```
